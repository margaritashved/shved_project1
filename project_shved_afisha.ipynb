#### Retention Rate

Retention Rate показывает, сколько пользователей из когорты относительно их изначального числа вернулось в последующие периоды.

```python
#Возьму дату, когда пользователи впервые проявили активность
first_ts_dates = visits.groupby('uid')['start_ts'].min()
first_ts_dates.name = 'first_ts_datetime'
first_ts_dates.head()
```

```python
#Объединяю с исходным датафреймом 
visits = visits.join(first_ts_dates,on='uid')
```

Когорты будут периодом 1 месяц, так как для данного сервиса это наиболее подходящий период.

```python
#Добавляю столбцы с первым месяцем когда произошло событие
visits['ts_month'] = visits['start_ts'].astype('datetime64[M]')
visits['first_ts_month'] = visits['first_ts_datetime'].astype('datetime64[M]')
#Для каждой строки датафрейма можно рассчитать lifetime пользователя в рамках когорты. 
#Использую константу np.timedelta64(1,'M'):
visits['cohort_lifetime'] = visits['ts_month'] - visits['first_ts_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype(int)
```

Чтобы найти Retention Rate, нужно изначальное число пользователей  в когорте разделить на число пользователей в каждый следующий месяц.

```python
cohorts_ts = visits.groupby(['first_ts_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
```

```python
# Проверка количества визитов
visits_pivot = visits.pivot_table(index='first_ts_month',columns='cohort_lifetime',values='uid',aggfunc='nunique')
display(visits_pivot.fillna(''))
# Нахожу исходное количество пользователей в когорте
inital_users_count = cohorts_ts[cohorts_ts['cohort_lifetime'] == 0][['first_ts_month','uid']]
inital_users_count = inital_users_count.rename(columns={'uid':'cohort_users'})
cohorts_ts = cohorts_ts.merge(inital_users_count,on='first_ts_month')
#расчет Retention Rate   
cohorts_ts['retention_rate'] = cohorts_ts['uid']/cohorts_ts['cohort_users']
retention_pivot = cohorts_ts.pivot_table(index='first_ts_month',columns='cohort_lifetime',values='retention_rate',aggfunc='sum')
```

```python
retention_pivot.mean() * 100
```

```python
#Создаю тепловую карту
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray',cmap="BuPu",yticklabels=retention_pivot.index.strftime('%Y-%m'),vmax=0.1);
plt.ylabel('first month', fontsize=14)
plt.xlabel('cohort lifetime', fontsize=14)
plt.show()
```

Retention Rate максимальный 8,5% для когорты 2017-09 , затем поступенно убывает и для когорты 2018-03 уже равен 4,2%. Больше всего на сервисе осталось пользователей когорты 2017-06, к концу периода осталось чуть больше половины пользователей. С каждой новой когортой поведение пользователей меняется. В целом пользователи, скорее всего, возвращаются только когда у них появляется потребность куда-либо сходить, мало кто остается на сервисе постоянно. Это связано с особенностями сервиса.

```python
print(f'Cредний Retention Rate на второй месяц «жизни» когорт {retention_pivot[1].mean():.2%}')
```

### Метрики электронной коммерции

#### Сколько времени проходит с момента первого посещения сайта до совершения покупки

Сколько времени проходит с начала сессии на сервисе до момента покупки.

```python
#Сгруппирую датафрейм в котором будет начало сессии пользователем
first_visit = visits.groupby('uid').agg({'start_ts':'min'})\
                .rename(columns={'start_ts':'first_ts_datetime'})\
                .reset_index()
```

```python
#Сгруппирую датафрейм в котором будет первая покупка пользователя
first_order = orders.groupby('uid').agg({'purchase_ts':'min'})\
                .rename(columns={'purchase_ts':'first_purchase_date'})\
                .reset_index()
```

```python
#Объеденяю 2 датафрейма в котором будет идентификатор пользователя, начало сессии и первая покупка
buying_time = first_visit.merge(first_order, on='uid', how='inner')
```

```python
#Считаю сколько времени в минутах проходит с момента начала сессии до момента покупки
buying_time['buying_time_min'] = (buying_time['first_purchase_date'] - buying_time['first_ts_datetime']).dt.seconds / 60
```

#### Cреднее количество покупок на одного покупателя за 6 месяцев

```python
#Сгруппирую датафрейм в котором будут первые покупки пользователей
first_orders = orders.groupby('uid').agg({'purchase_ts': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_ts']
```

```python
#Сгруппирую датафрейм в котором будут первые посещения пользователей
first_visits = visits.groupby('uid').agg({'start_ts': 'min'}).reset_index()
first_visits.columns = ['uid', 'first_visit_ts']
```

```python
#Создаю датафрейм с информацией о первых визитах и покупках пользователя, с информацией о первом месяцем покупки 
buyers = pd.merge(first_visits, first_orders, on='uid')
buyers['first_order_month'] = buyers['first_order_ts'].astype('datetime64[M]')
```

```python
#Сгруппирую датафрейм с датой первого заказа,суммой заказа
first_order_dates = orders.groupby('uid')['purchase_ts'].min()
first_order_dates.name = 'first_order_datetime'
```

```python
#Создаю единый датафрейм с полной информацией о первых покупках пользователей
orders = orders.join(first_order_dates, on='uid', how='left', lsuffix='_left')
```

```python
#В датафрейм orders добавляю информацию о месяце покупки и первом месяце покупки
orders['order_month'] = orders['purchase_ts'].astype('datetime64[M]')
orders['first_order_month'] = orders['first_order_datetime'].astype('datetime64[M]')
```

```python
#Считаю количество покупателей в месяц
cohort_sizes = buyers.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.rename(columns={'uid': 'n_buyers'}, inplace=True)
```

```python
#Датафрейм с информацией о когортах
cohorts = pd.merge(orders, buyers, how='inner', on=['uid', 'first_order_month'])\
    .groupby(['first_order_month', 'order_month'])\
    .agg({'revenue': ['sum', 'count']}).reset_index()
```

```python
#Время жизни когорты
cohorts['age_month'] = ((cohorts['order_month'] - cohorts['first_order_month']) / np.timedelta64(1,'M')).round()
cohorts.columns = ['first_order_month', 'order_month', 'revenue', 'n_orders', 'age_month']
```

```python
cohorts_report = pd.merge(cohort_sizes, cohorts, on = 'first_order_month')
#Сумма заказа пользователя
cohorts_report['rev_per_buyer'] = cohorts_report['revenue'] / cohorts_report['n_buyers']
#Количество покупок пользователя
cohorts_report['orders_per_buyer'] = cohorts_report['n_orders'] / cohorts_report['n_buyers']
```

```python
#Когортный отчет для того, чтобы посмотреть сколько в среднем пользователь делает покупок за пол года месяц
cohorts_age = cohorts_report.pivot_table(
index='first_order_month', 
columns='age_month', 
values=['rev_per_buyer', 'orders_per_buyer'], 
aggfunc='sum'
).cumsum(axis=1)
```

```python
print('В среднем за 6 месяцев пользователь делает {:.2f} заказа.'.format(cohorts_age['orders_per_buyer'][5].mean()))
```

#### "Пожизненная" ценность клиента (LTV)

```python
#Маржинальность сервиса — 100%.
margin_rate = 1
```

```python
#Информация о первом месяце покупки, использую предыдущий датафрейм first_orders, добавляю столбец месяц покупки
first_orders['first_order_ts']=first_orders['first_order_ts'].astype('datetime64[M]')
```

```python
#Информация о количестве новых покупателей за каждый месяц
cohort_sizes.head()
```

```python
#Группирую таблицу покупок в которой первй месяц покупки,месяц заказа и сумма выручки
cohorts_new = orders.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
display(cohorts_new.head())
```

```python
#Объединяю 2 датафрейма
cohorts_all = pd.merge(cohort_sizes, cohorts_new, on='first_order_month')
display(cohorts_all.head())
```

При расчете LTV необходимо учитывать валовую прибыль. Валовая прибыль равна выручке умноженной на маржинальность. LTV — относительный показатель, его удобнее смотреть по «возрастным» когортам, где в столбцах — не месяц заказа, а возраст (age) когорты.

```python
#Валовая прибыль
cohorts_all['gp'] = cohorts_all['revenue'] * margin_rate
```

```python
#Длительность жизни когорты
cohorts_all['age'] = (cohorts_all['order_month'] - cohorts_all['first_order_month']) / np.timedelta64(1, 'M')
cohorts_all['age'] = cohorts_all['age'].round().astype('int')
```

```python
#LTV = валовая прибыль за месяц /число пользователей в каждой когорте
cohorts_all['ltv'] = cohorts_all['gp'] / cohorts_all['n_buyers']
```

```python
cohorts_all.head()
```

```python
output = cohorts_all.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean').round(2).cumsum(axis=1)

cohorts_all.fillna('')

sns.set(style='white')
plt.figure(figsize=(15, 9))
plt.title('LTV', fontsize=20)
sns.heatmap(output, cmap="BuPu", annot=True, fmt='.3', linewidths=1, linecolor='gray', yticklabels=output.index.strftime('%Y-%m'),)
plt.xlabel('age', fontsize=14)
plt.ylabel('first_order_month', fontsize=14)
plt.show()
```

```python
print('Средний LTV за 6 месяцев после первой покупки: {:.2f}.'.format(output.mean(axis=0)[5]))
```

```python
#Выделяются 1 и 4 когорты
print('В среднем каждый покупатель из первой когорты принес {:.0f} у.е. валовой прибыли за 11 месяцев «жизни».'.format(output.loc['2017-06-01'].sum()))
print('В среднем каждый покупатель из четвертой когорты принес {:.0f} у.е. валовой прибыли за 11 месяцев «жизни».'.format(output.loc['2017-09-01'].sum()))
```

Наибольшую прибыль принесли пользователи июньской и сентябрьской когорт. Лето пора фестивалей, возможно, это сыграло свою роль, а в сентябре пользователи начинают планировать осенние и зимние вечера.

#### Окупаемость расходов (ROI)

```python
#Из датафрейма orders забираю столбцы сумма заказа и идентификатор пользователя
orders_roi = orders[['revenue', 'uid']]
#Из датафрейма visits забираю столбцы идентификатор источника перехода на сайт,идентификатор пользователя и дата первого посещения
visits_roi = visits[['source_id', 'uid', 'first_ts_datetime']]
```

```python
#Пользователь мог прийти на сайт из разных источников
#Первый заход и удаление дубликатов
visits_roi = visits_roi.groupby(['uid', 'source_id'], as_index=False)['first_ts_datetime'].min()
visits_roi = visits_roi.drop_duplicates(subset=['uid','first_ts_datetime'])
```

```python
#Объединяю датафреймы по столбцу uid
roi_table = pd.merge(orders_roi, visits_roi, on='uid')

#Доходы по источнику
revenue_by_source = roi_table.groupby('source_id', as_index=False).agg({'revenue': 'sum'})

#Расходы по источнику
costs_roi = costs.groupby('source_id', as_index=False)['costs'].sum()

roi_total = revenue_by_source.merge(costs_roi, on='source_id')

#ROI - Отношение заработанных денег к потраченным в %
roi_total['roi_%'] = ((roi_total['revenue'] - roi_total['costs'])/ roi_total['costs']*100).round(2)

display(roi_total)
print('Сумма заказов:', (roi_total['revenue'].sum()).round())
print('Расходы:', (roi_total['costs'].sum()).round())
```

Если ROI меньше 100%, то компания несет убытки, если равен 100% — вложения окупились, но прибыли нет. Если больше 100%, то компания получает прибыль. В данном случае при переходе на сайт с 1 источника вложения в маркетинг более чем окупились, почти на 400%. Что касается остальных источников, то вложения в маркетинг не окупились.

```python
print('Общий ROI' , ((roi_total['revenue'].sum() - roi_total['costs']).sum()/ roi_total['costs'].sum()*100).round(0) , '%')
```
